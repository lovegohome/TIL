# Target 변수 & Features

- Target 변수가 2개 있다고 하면, 각각 타겟으로 두고 나머지를 feature로 둘 때 학습데이터 set이 어떻게 되는지
- 시계열 분석을 하면 유니베리어트, 즉 변수가 하나
- 변수 중 하나를 타겟으로 놔버리면, 나머지는 input (통계로는 독립변수, 여기서는 Parameter or Feature라고 한다. 
- Feature를 갖고 feature extraction(특징추출), 차원축소, 학습데이터 셋을 어떻게 만드는지 학습을 하는 중
- Feature들 간의 상관관계를 보고, Feature들과 Target variable의 인과관계를 볼 거다.
- (input) 상관관계를 보는 데 있어서 이 변수들이 하나일 적에 univariate, 여러 개의 변수가 있으면 multivariate(다변량)
- 변수들 간의 관계를 보기 위해 공간에서의 위치 및 분포를 같이 봤었다. (구분하기 위해)
  - 일차원 : 평균 
  - 2개: 분산 
  - 여러개 : covariance(공분산) 
  - 확률변수: 확률로 나타냄(평균, 분산, 공분산)
    - tip, size 각각의 평균, 분산
      tip과 size가 (-)기울기를 가진 그래프를 보이면 공분산 값이 0보다 큰 값을 갖게 된다. 
    - 공분산 = 0 : 서로 완전히 독립
- 1개를 목표변수, 나머지(7개)를 인풋변수. 
  그러면 7개 차원으로 공분산 계산, 그리고 거기서 [Eigenvalue, 고유값 cf.요인분석]를 구해서 순서대로 세우면 7개 차원에서의 **determinant**(행렬식)이 나오는데, 그 행렬식의 크기에 따라서 cut-off를 할 수 있었다. 가장 모델에 영향력 주는 변수들 파악하고 나머지는 버린다. = PCA 
  PCA = **Principal component analysis**, 주성분 분석

- 즉, 변수들간의 관계를 확률분포값으로 나타낸다는 것이다.

### AARRR

- Acquisition (취득, 사용자 획득) 
  : 사용자가 어떻게 서비스를 접하는가?
- Activation (사용자 활성화)
  : 사용자가 처음 서비스를 이용했을 때 경험이 좋았는가?
- Retention (보유, 사용자 유지 )
  :사용자가 우리 서비스를 계속 이용하는가?
- Revenue (매출)
  : 어떻게 돈을 버는가?
- Referral (추천)
  : 사용자가 다른 사람들에게 제품을 소개하는가?
