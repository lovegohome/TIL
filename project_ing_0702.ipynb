```python
def outliers_iqr(data):
    data1=data.iloc[:,3]
    print(data1)
    q1, q3 = np.percentile(data1,[2,75])
    iqr = q3 - q1
    lower_bound = q1 - (iqr * 1.5)
    upper_bound = q3 + (iqr * 1.5)
    result_ind=np.where((data1 > upper_bound)| (data1 < lower_bound))
#     print(result_ind)
    remoevData= pd.DataFrame(data.iloc[result_ind[0],3])
    return remoevData
```
