* 주로 사용해볼 함수는 relplot()

이 함수는 scatter plot과 line plot을 사용해서 plotting할 수 있으며, FacetGird와 axes-level function인 scatterplot()과 lineplot() 중에 하나를 결합해서 그래프를 그리게 됩니다.

  relplot을 사용하고 매개변수 kind="scatter"나 kind="line"을 통해서 그래프 종류를 선택할 수 있고, kind 매개변수를 지정하지 않으면 scatter가 기본으로 적용됩니다. 

  색상이나 추가적인 커스터마이징이 많이 필요한 경우에 matplotlib의 메소드와 함께 사용하면 될 것 같습니다.



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
#sns.set_theme()
```

## **1. scatter plots**

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

data는 사용되는 dataset입니다. 사용할 수 있는 dataset으로는 pandas.DataFrame, numpy.ndarray, mapping or sequence가 가능합니다.(참고로 tips의 type은 pandas의 DataFrame입니다)

```python
type(tips)
```

```python
sns.relplot(x="total_bill", y="tip", data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", hue="sex", data=tips)
```

```python
#sns.relplot(x="total_bill", y="tip", hue="smoker", style="smoker", data=tips)
#sns.relplot(x="total_bill", y="tip", hue="smoker", style="sex", data=tips)
sns.relplot(x="total_bill", y="tip", hue="smoker", style="time", data=tips)
```

* hue semantic

  smoker, sex, time = 카테고리(qualitative palette 적용)

  numeric(수치)일 경우 : sequential palette(color palette)

```python
sns.relplot(x="total_bill", y="tip", hue="size", data=tips)
```

palette 매개변수에 color_palette()를 통해 string으로 값을 지정해서 색상을 지정

(커스터마이징하기 시작하면 조금 복잡해지기 때문에, 기본으로 사용 추천)

```python
#sns.relplot(x="total_bill", y="tip", hue="size", palette="ch:r=4,l=.7", data=tips)
sns.relplot(x="total_bill", y="tip", hue="size", palette="ch:r=-.5,l=.75", data=tips)
```

```python
#sns.relplot(x="total_bill", y="tip", hue="size", data=tips)
#sns.relplot(x="total_bill", y="tip", hue="size", size="size", data=tips)
#sns.relplot(x="total_bill", y="tip", size="size", data=tips)
#sns.relplot(x="total_bill", y="tip", size="tip", data=tips)
#sns.relplot(x="total_bill", y="tip", hue="size", size="tip", data=tips)
sns.relplot(x="total_bill", y="tip", size="tip", sizes=(10,200), data=tips)
```
