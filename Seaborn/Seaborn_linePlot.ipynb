```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#sns.set_theme(style="darkgrid")
sns.set_theme()
```

## 2. line plots

* 연속된 값을 가진 변수들의 시각화에 유용
* lineplot(), sns.relplot(kind="line")


먼저 line plot을 위한 dataset을 생성.

0~500초 값의 분포

```python
df = pd.DataFrame(dict(time=np.arange(500), 
                       value=np.random.randn(500).cumsum()))
```

```python
df.head()
```

```python
g = sns.relplot(x="time", y="value", kind="line", data=df)
#g = sns.relplot(x="time", y="value", data=df)
g.fig.autofmt_xdate()
```

lineplot()은 기본적으로 그래프를 그리기 전에 x값을 기준으로 데이터를 정렬하고 그래프를 그립니다. 따라서, x가 정렬되지 않은 데이터를 가지고 그래프를 그리면 어떻게 될까.

```python
df = pd.DataFrame(np.random.randn(500, 2).cumsum(axis=0), columns=["x", "y"])
#df = pd.DataFrame(np.random.randn(500, 2).cumsum(), columns=["x", "y"])
```

```python
df.head()
```

```python
sns.relplot(x="x", y="y", kind="line", data=df)
```

이렇게 x가 순차적이지 않음에도 불구하고, line이 정렬된 x의 값들로 이어져있는 것을 볼 수 있습니다. 
sort=False 매개변수로 정렬을 적용하지 않을 수 있는데, 정렬을 적용하지 않으면 정렬되지 않은 x의 값을 순차적으로 이어버리게 됩니다.

```python
#sns.relplot(x="x", y="y", sort=False, kind="line", data=df)
sns.relplot(x="x", y="y", kind="line", data=df, sort=False)
```

더 복잡한 dataset은 하나의 x 변수에 대해서 여러 개의 측정값 y를 가질 수 있습니다. 이러한 dataset인 경우에 seaborn의 기본 동작은 각 x의 측정값들에 대해서 평균과 95%의 신뢰구간을 평균 주위에 표시하게 됩니다.
